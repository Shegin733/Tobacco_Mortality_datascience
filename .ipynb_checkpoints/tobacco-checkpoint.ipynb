```python
import numpy as np

# Loop through each row with NaNs in columns 9-13
for index, row in df[df.iloc[:, 9:14].isna().any(axis=1)].iterrows():
    year = row['Year']  # Current year
    icd10_code = row['ICD10_Code']
    icd10_diagnosis = row['ICD10_Diagnosis']

    # Find matching rows in the previous and next years
    prev_year_row = df[(df['Year'] == year - 1) & 
                       (df['ICD10_Code'] == icd10_code) & 
                       (df['ICD10_Diagnosis'] == icd10_diagnosis)]
    
    next_year_row = df[(df['Year'] == year + 1) & 
                       (df['ICD10_Code'] == icd10_code) & 
                       (df['ICD10_Diagnosis'] == icd10_diagnosis)]
    
    # Check if both rows are found
    if not prev_year_row.empty and not next_year_row.empty:
        print(f"\nProcessing row index {index}, ICD10_Code: {icd10_code}, ICD10_Diagnosis: {icd10_diagnosis}")
        print(f"Previous Year Row (Year {year-1}):\n{prev_year_row.iloc[0, 9:14]}")
        print(f"Next Year Row (Year {year+1}):\n{next_year_row.iloc[0, 9:14]}")
        
        # For each column 9-13, calculate mean and fill if necessary
        for col in range(9, 14):
            prev_value = prev_year_row.iloc[0, col]
            next_value = next_year_row.iloc[0, col]

            # Print extracted values
            print(f"Column {df.columns[col]} - Prev Value: {prev_value}, Next Value: {next_value}")

            # Check if both values are non-NaN
            if not np.isnan(prev_value) and not np.isnan(next_value):
                mean_value = (prev_value + next_value) / 2
                df.at[index, df.columns[col]] = mean_value
                print(f"Filled NaN at index {index}, column {df.columns[col]} with mean value: {mean_value}")
            elif not np.isnan(prev_value):
                df.at[index, df.columns[col]] = prev_value
                print(f"Filled NaN at index {index}, column {df.columns[col]} with prev year value: {prev_value}")
            elif not np.isnan(next_value):
                df.at[index, df.columns[col]] = next_value
                print(f"Filled NaN at index {index}, column {df.columns[col]} with next year value: {next_value}")
            else:
                print(f"Cannot fill NaN at index {index}, column {df.columns[col]} due to missing values in both years.")
    
    else:
        print(f"Could not find matching rows for index {index}, ICD10_Code: {icd10_code}, ICD10_Diagnosis: {icd10_diagnosis}")
```
