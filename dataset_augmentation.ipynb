```python
import pandas as pd

# Load the datasets
admissions = pd.read_csv('admissions.csv')
fatalities = pd.read_csv('fatalities.csv')

# Function to expand the year column in admissions DataFrame
def expand_year(year):
    try:
        start_year, end_year = year.split('/')
        return [int(start_year), int(end_year)]
    except Exception as e:
        print(f"Error processing year '{year}': {e}")
        return []

# Expand the year column in admissions
admissions['Year'] = admissions['Year'].apply(expand_year)

# Create a new DataFrame to hold expanded rows with only ICD10 code, Year, and Value
expanded_admissions = []
for index, row in admissions.iterrows():
    for year in row['Year']:
        expanded_admissions.append({
            'ICD10 Code': row['ICD10 Code'],
            'Year': year,
            'Value': row['Value']
        })

# Convert expanded list back to DataFrame
expanded_admissions_df = pd.DataFrame(expanded_admissions)

# Merge the expanded admissions DataFrame with fatalities DataFrame
merged_df = pd.merge(fatalities, expanded_admissions_df[['ICD10 Code', 'Year', 'Value']], 
                     how='left', on=['ICD10 Code', 'Year'])

# Rename the Value column from admissions to admissions
merged_df.rename(columns={'Value_y': 'admissions'}, inplace=True)  # Change 'Value' to 'Value_y'

# Print merged DataFrame columns to debug
print("Columns in merged_df:")
print(merged_df.columns)

# Select only the necessary columns from merged_df
try:
    final_df = merged_df[['ICD10 Code', 'Year', 'admissions']]  # Keep only these columns
except KeyError as e:
    print(f"Error selecting columns: {e}")
    # Display the merged DataFrame for further inspection
    print("Merged DataFrame:")
    print(merged_df.head())

# Display the result
merged_df.rename(columns={'Value_x': 'Deaths'}, inplace=True)
print("\nFinal Merged Data:")
print(final_df.head())

# Optionally save the result to a new CSV file
# final_df.to_csv('final_fatalities_with_admissions.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats


merged_df['admissions'] = pd.to_numeric(merged_df['admissions'], errors='coerce')

# Define a function to remove outliers using the IQR method
def remove_outliers_iqr(data, column):
    """Removes outliers from a DataFrame column using the IQR method."""
    if pd.api.types.is_numeric_dtype(data[column]):
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    else:
        print(f"Skipping column '{column}' as it is not numeric.")
        return data

# Define a function to remove outliers using the Z-score method
def remove_outliers_zscore(data, column, threshold=3):
    """Removes outliers from a DataFrame column using the Z-score method."""
    if pd.api.types.is_numeric_dtype(data[column]):
        z_scores = np.abs(stats.zscore(data[column]))
        return data[z_scores < threshold]
    else:
        print(f"Skipping column '{column}' as it is not numeric.")
        return data

# List of columns to check for outliers
columns_to_check = [
    
    'admissions',
    'Therapy_Cost',
    
    'Real_Households_Income',
   
    'Household_Expenditure_Tobacco',
    'Addicts(Age Group)'
]

# Remove outliers for each specified column using IQR and Z-score methods
for column in columns_to_check:
    # Apply IQR method
    merged_df = remove_outliers_iqr(merged_df, column)
    
    # Apply Z-score method (optional, for more precision)
    merged_df = remove_outliers_zscore(merged_df, column)

# Check the shape after removing outliers
print(f"Shape after removing outliers: {merged_df.shape}")

# Save the cleaned dataset


# Plotting to visualize outliers removal (boxplot)
for column in columns_to_check:
    plt.figure(figsize=(10, 6))
    sns.boxplot(merged_df[column])
    plt.title(f'Boxplot of {column} After Outlier Removal')
    plt.show()
```

```python



# Convert 'Deaths' and 'admissions' to numeric, coercing errors to NaN
merged_df['Deaths'] = pd.to_numeric(merged_df['Deaths'], errors='coerce')
merged_df['admissions'] = pd.to_numeric(merged_df['admissions'], errors='coerce')

def remove_outliers_iqr(data, column):
    """Removes outliers from a DataFrame column using the IQR method."""
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

def remove_outliers_z_score(data, column, threshold=3):
    """Removes outliers from a DataFrame column using the Z-score method."""
    z_scores = np.abs(stats.zscore(data[column].dropna()))  # Calculate Z-scores, drop NaN
    return data[(z_scores < threshold) | (data[column].isna())]  # Keep only inliers

def remove_outliers_modified_z_score(data, column, threshold=3.5):
    """Removes outliers from a DataFrame column using the Modified Z-score method."""
    median = data[column].median()
    mad = np.median(np.abs(data[column] - median))  # Median Absolute Deviation
    modified_z_scores = 0.6745 * (data[column] - median) / mad  # Modified Z-score formula
    return data[(np.abs(modified_z_scores) < threshold) | (data[column].isna())]  # Keep only inliers

def winsorize_data(data, column, lower_percentile=0.01, upper_percentile=0.99):
    """Winsorizes the data by capping extreme values."""
    lower_bound = data[column].quantile(lower_percentile)
    upper_bound = data[column].quantile(upper_percentile)
    data[column] = np.clip(data[column], lower_bound, upper_bound)
    return data

# Log transformation function
def log_transform(data, column):
    """Applies log transformation to a DataFrame column."""
    data[column] = np.log1p(data[column])  # Use log1p to avoid log(0)
    return data

# Check the distributions before removing outliers
for column in ['Deaths', 'admissions']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=merged_df, x=column)
    plt.title(f'Box Plot of {column} (Before Outlier Removal)')
    plt.show()

# Remove outliers using IQR method first
for column in ['Deaths', 'admissions']:
    merged_df = remove_outliers_iqr(merged_df, column)

# Apply Z-Score method
for column in ['Deaths', 'admissions']:
    merged_df = remove_outliers_z_score(merged_df, column)

# Apply Modified Z-Score method
for column in ['Deaths', 'admissions']:
    merged_df = remove_outliers_modified_z_score(merged_df, column)

# Apply Winsorization to both columns
for column in ['Deaths', 'admissions']:
    merged_df = winsorize_data(merged_df, column)

# Apply log transformation to Deaths
merged_df = log_transform(merged_df, 'Deaths')
merged_df = merged_df[(merged_df['Deaths'] > 6)]  


# Check the shape after removing outliers
print(f"Shape after removing outliers: {merged_df.shape}")

# Check the distributions after removing outliers
for column in ['Deaths', 'admissions']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=merged_df, x=column)
    plt.title(f'Box Plot of {column} (After Combined Outlier Removal)')
    plt.show()



```
